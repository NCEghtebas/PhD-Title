```python
import nltk

def edit_distance(word, word_count_dict):
    diff = []
    for i in word_count_dict:
        if i != word:
            edit_distance = nltk.edit_distance(i, word)
            diff.append(edit_distance)
    return diff

diff_rethink = edit_distance("Rethinking", word_count_clean)
# avg, min_v, max_V, mode = basic_stats(diff_rethink)
# 
```

```python
combs = itertools.combinations(word_count_clean,2)


def most_similar(word, word_count_dict):
    diff = []
    words = {}
    for i in word_count_dict:
        if i != word:
            edit_distance = nltk.edit_distance(i, word)
            diff.append(edit_distance)
            if edit_distance <= min(diff):
                words[i] = edit_distance     
    most_similar = [k for k,v in words.items() if v==min(diff)]
    return most_similar
    
most_similar("Rethinking", word_count_clean)
```

```python
similar_words = {}
for i in word_count_clean:
    similar_words[i] = most_similar(i, word_count_clean)

pp.pprint(similar_words)
```
